```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("theta_final.csv")
thetas = df["theta"].values

plt.hist(thetas, bins=50, density=True, alpha=0.7)
plt.xlabel("Î¸  (radians)")
plt.ylabel("Probability density")
plt.title("Histogram of final U(1) holonomy angles")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import gaussian_kde

kde = gaussian_kde(thetas)
x = np.linspace(-np.pi, np.pi, 500)
plt.plot(x, kde(x), lw=2)
plt.xlabel("Î¸  (radians)")
plt.title("Kernel density of Î¸")
plt.grid(True)
plt.show()
```
